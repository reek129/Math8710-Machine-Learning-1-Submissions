```python
from sklearn.datasets import load_wine,load_iris

import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
flower =load_iris()
X= flower.data
y = flower.target

X_train, X_test, y_train, y_test = train_test_split(flower.data, flower.target, test_size=0.33, random_state=42)
```

# Decision Tree

```python
clf = DecisionTreeClassifier(random_state=0,max_depth=5)
# clf = clf.fit(X_train,y_train )
clf = clf.fit(X,y )

fig = plt.figure(figsize=(15, 10))
plot_tree(clf,
         feature_names=flower.feature_names,
         class_names=flower.target_names,
         filled=True, impurity=True, 
          rounded=True)
fig.savefig("rf_iris3/"+"DecisionTree2"+'.png')
```

# Random Forest

```python

rf = RandomForestClassifier(n_estimators=100, 
                            max_depth=3,
                            max_features='auto', 
                            min_samples_leaf=4,
                            bootstrap=True, 
                            n_jobs=-1, 
                            random_state=0)
rf.fit(X_train, y_train)

index = len(rf.estimators_)

for i in range(0,index):
    fig = plt.figure(figsize=(15, 10))
    plot_tree(rf.estimators_[i], 
          feature_names=flower.feature_names,
          class_names=flower.target_names, 
          filled=True, impurity=True, 
          rounded=True)
    fig.savefig("rf_iris3/"+str(i)+'.png')
```

# Accuracy for decision tree

```python
y_dt = clf.predict(X_test)
print("Confusion Matrix")
print(confusion_matrix(y_test, y_dt))
print("accuracy")
```

```python
y_rf = rf.predict(X_test)
confusion_matrix(y_test, y_rf)
```
